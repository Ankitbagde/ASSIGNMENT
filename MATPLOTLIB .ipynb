```python
(#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
The subplot() function is used to create multiple plots within a single figure. It allows you to display different graphs side-by-side or stacked on top of each other.
This can be useful for comparing different data sets or visualizing different aspects of the same data set.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot. 
    The index starts at 1 and increases from left to right and top to bottom.
    For example, if you want to create a 2x2 grid of subplots, the first subplot would have an index of 1, the second subplot would have an index of 2, and so on.

Here are four line plots created using the subplot() function:)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot each line on a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Set the title for each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Set the x and y axis labels for each subplot
axs[0, 0].set_xlabel('X axis')
axs[0, 0].set_ylabel('Y axis')
axs[0, 1].set_xlabel('X axis')
axs[0, 1].set_ylabel('Y axis')
axs[1, 0].set_xlabel('X axis')
axs[1, 0].set_ylabel('Y axis')
axs[1, 1].set_xlabel('X axis')
axs[1, 1].set_ylabel('Y axis')

# Display the plot
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A bar plot is a graphical representation of data that uses rectangular bars to represent the values of the data points. The length or height of the bars corresponds to the magnitude of the data. It is used to compare different categories of data, where each category is represented by a separate bar.

A vertical bar plot displays the categories along the x-axis and the values along the y-axis. 
A horizontal bar plot displays the categories along the y-axis and the values along the x-axis.

Here's how to create a vertical bar plot and a horizontal bar plot using the provided data:
The bar() function is used to create a vertical bar plot and the barh() function is used to create a horizontal bar plot. 
The title(), xlabel(), and ylabel() functions are used to add labels to the plot. The show() function is used to display the plot.

The output of this code will be two separate plots: a vertical bar plot and a horizontal bar plot.
    The vertical bar plot will display the profit values for each company along the x-axis, with a separate bar for each company.
    The horizontal bar plot will display the profit values for each company along the y-axis, with a separate bar for each company.
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

### Vertical bar plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit (RS)')
plt.show()

## Horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit (RS)')
plt.ylabel('Company')
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
A box plot is a graphical representation of data that shows the distribution of a dataset by displaying the median, 
quartiles, and outliers of the data. The box in the plot represents the middle 50% of the data, 
with the lower and upper boundaries of the box representing the first and third quartiles of the data. The line inside the box represents the median of the data.
The "whiskers" extend from the box to show the range of the data, excluding any outliers that are plotted as individual points.

Box plots are used to compare the distributions of different datasets or to visualize the spread and skewness of a single dataset.

To create a box plot using the provided data, we can use the boxplot() function from Matplotlib:


The boxplot() function takes in a list of data arrays as input and generates a box plot for each one. We create a list data that contains the two data arrays box1 and box2, and a list labels that contains the labels for each box.
We then pass these lists to the boxplot() function, and use the title() and ylabel() functions to add labels to the plot.

The output of this code will be a box plot that displays the distributions of the two datasets side-by-side, with the labels "Box 1" and "Box 2" along the x-axis. The median, quartiles, and whiskers of each dataset are displayed in the plot. 
Any outliers that fall outside the range of the whiskers are plotted as individual points.
```

```python
##python code
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```
