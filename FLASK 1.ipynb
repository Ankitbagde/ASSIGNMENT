```python
Q3. What is App routing in Flask? Why do we use app routes?
```

```python
In Flask, app routing is the process of defining URL patterns (or routes) for an application.
It allows us to map URLs to view functions that generate responses for the requests made to the server.

To define a route in Flask, we use the @app.route() decorator, followed by the URL pattern. 
The decorator tells Flask what URL should trigger the associated function.

For example, to define a route for the home page of a website, we can use the following code:

python code
@app.route('/')
def home():
    return 'Welcome to my website!'


In this example, the @app.route() decorator is used to define the URL pattern '/'. 
The associated function home() returns the welcome message when the home page is accessed.

We use app routes in Flask to create a well-organized and structured application. By using routes,
we can easily map each URL to a specific function that handles the corresponding request. This makes 
it easier to maintain the code and add new features to the application.

Moreover, using app routes in Flask allows us to create a RESTful API, which is a popular architectural
style for designing web-based applications. In RESTful APIs, each route represents a resource that can be accessed through HTTP methods such as GET, POST, PUT, and DELETE. This makes it easier to develop and maintain web-based applications that follow a standard architecture.
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

```python
The url_for() function is used in Flask for URL building. It generates a URL to a specific function, based on the name of the function and any arguments provided.
In the above example, we define two routes, one for the homepage (/) and one for a user profile (/user/<username>). We then use the url_for() function to generate URLs for these routes.

In the with app.test_request_context() block, we call the url_for() function twice. The first call generates a URL for the index function (which maps to the homepage), and the second call generates a URL for the user_profile function, with the username John.

The output of the print() statements shows the generated URLs: / for the homepage and /user/John for the user profile with the username John.
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the homepage.'

@app.route('/user/<username>')
def user_profile(username):
    return 'Welcome to your profile, {}!'.format(username)

with app.test_request_context():
    print(url_for('index'))
    # Output: '/'
    print(url_for('user_profile', username='John'))
    # Output: '/user/John'
```
